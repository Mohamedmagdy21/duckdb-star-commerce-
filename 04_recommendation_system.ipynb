### Scoring Factors

| category_score | 0.15 | Same category = 1.0, same parent = 0.5, else 0.0 |

| popularity_score | 0.10 | Total units sold |
| lift_score | 0.15 | P(A and B) / (P(A) x P(B)), statistical association |
